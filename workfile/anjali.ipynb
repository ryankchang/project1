```python

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import datetime
import pprint
import csv
from datetime import datetime
```

```python
county_data_file= 'data/LOS_ANGELES_COUNTY_RESTAURANT_AND_MARKET_INSPECTIONS.csv'
```

```python
county_data_file=pd.read_csv(county_data_file)
county_data_file.head()
```

```python
county_data_file.count()
```

```python
columns=['ACTIVITY DATE','FACILITY ID','FACILITY NAME','PROGRAM STATUS','FACILITY ADDRESS','FACILITY CITY','FACILITY ZIP','SCORE','GRADE']
new_data_county=county_data_file[columns]
new_data_county['ACTIVITY DATE']=pd.to_datetime(new_data_county['ACTIVITY DATE'])
new_data_county.head()
```

```python
new_data_county.sort_values(by=["ACTIVITY DATE","FACILITY ID","FACILITY ADDRESS"],inplace = True,ascending=True)
new_data_county.head()
```

```python
unique_county_data = new_data_county.drop_duplicates(subset = ['FACILITY ID'], keep = 'last')
unique_county_data.count()
```

```python
unique_county_data = unique_county_data.loc[(unique_county_data['PROGRAM STATUS'] != "INACTIVE"),:]
unique_county_data.count()
```

```python
unique_county_data = unique_county_data.dropna(how='any')
unique_county_data.count()
```
